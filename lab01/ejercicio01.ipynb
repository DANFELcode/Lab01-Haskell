¿Qué significa que Haskell sea un lenguaje fuertemente tipado y evaluado perezosamente?


Significa que los tipos deben estar bien definidos en haskell, y no se permiten operaciones entre tipos incompatibles, y es de evaluación perezosa porque no ejecuta un caso a no ser que sea estrictamente necesario o se le indique

```haskell
doble :: Int -> Int
doble x = x*2
```

```haskell
doble 6
```

```haskell
esMayorEdad :: Int -> Bool
esMayorEdad x = x > 18
```

```haskell
esMayorEdad 20
```

```haskell
areaCirculo :: Float -> Float
areaCirculo r = pi * r * r
```

```haskell
areaCirculo 19
```

¿Qué diferencias encuentras entre los tipos Int y Float?

Int es un tipo de dato para enteros y Float es un tipo de dato para representar numeros de punto flotante
