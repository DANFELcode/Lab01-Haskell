yo creo que lo que hara sera multiplicar los elementos de una lista indefinida desde 1 por 2, luego tomara los 5 primeros y esto funcionara gracias a la evaluzación peresoza de haskell

```haskell
take 5 (map (*2) [1..])
```

efectivamente tal como lo predije
