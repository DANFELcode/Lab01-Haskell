```haskell
head [10,20,30,40]
tail [10,20,30,40]
take 3 [1..10]
drop 4 [1..10]
length [1,2,3,4,5]
```

Head retorna la cabeza de la lista, tail retorna la cola de la lista(todos menos el primero), take toma 3 elementos de la lista, drop guarda la lista borrando los 4 primeros elementos, length calcula la cantidad de elementos de la lista

¿Qué hace la expresión [1..10] y cómo se relaciona con la evaluación perezosa?

La expresion representa una lista de 1 a 10 es de evaluzación peresoza puesto que haskell no necesita leer todos los numeros de la lista para descifrar cual es


```haskell
primerElemento :: [a] -> a
primerElemento (x:_) = x
primerElemento [14,4,6,7]
```

```haskell
productoLista :: [Int] -> Int
productoLista [] = 1
productoLista [x] = x
productoLista l = head l * productoLista (tail l)
```

```haskell
productoLista [1,5,8,9]
```

```haskell
eliminarNegativos :: [Int] -> [Int]
eliminarNegativos [x]  
    |x > 0 = [x]
    |x < 0 = []
    |otherwise = [0]
eliminarNegativos (x:xs) = eliminarNegativos [x] ++ eliminarNegativos xs
```

```haskell
eliminarNegativos [0]
```

```haskell
invertir :: [a] -> [a]
invertir [] = []
invertir [x] = [x]
invertir (x:xs) = invertir 
```

```haskell
invertir [1,3,5]
```
