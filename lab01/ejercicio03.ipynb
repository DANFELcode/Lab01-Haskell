```haskell
suma:: Int -> Int -> Int
suma x y = x + y
```

```haskell
cuadrado:: Int -> Int
cuadrado x  = x*x 
```

```haskell
sumaCuadrados :: Int -> Int -> Int
sumaCuadrados x y = suma (cuadrado x) (cuadrado y)
```

```haskell
sumaCuadrados 2 2
```

```haskell
duplicar :: Int -> Int
duplicar x = x * 2

dobleLista :: [Int] -> [Int]
dobleLista xs = map duplicar xs
```

```haskell
clasificarNota :: Int -> String
clasificarNota n
    | n < 3 = "Reprobado"
    | n < 4 = "En recuperacion"
    | otherwise = "Aprobado"
```

```haskell
clasificarNota 2
```

```haskell
fibonacci :: Int -> Int
fibonacci 0 = 0
fibonacci 1 = 1
fibonacci n = fibonacci (n-1) + fibonacci (n-2)
```
